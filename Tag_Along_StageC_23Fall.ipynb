```python
from sklearn.preprocessing import OneHotEncoder

categorical_columns = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService','OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies','Contract', 'PaperlessBilling', 'PaymentMethod']

df_categorical = categorical[categorical_columns]

# Initialize OneHotEncoder
encoder = OneHotEncoder(sparse=False)

# Fit and transform the selected columns
encoded_data = encoder.fit_transform(df_categorical)

# Get feature names after encoding
feature_names = encoder.get_feature_names_out(categorical_columns)

# Convert the encoded data back to a DataFrame
encoded_df = pd.DataFrame(encoded_data, columns=feature_names)


# Concatenate the encoded DataFrame with the original DataFrame, dropping the original categorical columns
categorical = pd.concat([categorical.drop(categorical_columns, axis=1), encoded_df], axis=1)

```

# Modelling

## Random Forest and Extra Tree Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score

rf_classifier = RandomForestClassifier(n_estimators=100, random_state=1)

rf_classifier.fit(X_train, y_train)

predictions = rf_classifier.predict(X_test)

accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {accuracy}")
```

```python
predictions
```

```python
pred_frame = pd.DataFrame(predictions,columns=["Predicted Churn"])
pred_frame
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.datasets import make_classification

et_classifier = ExtraTreesClassifier(n_estimators=100, random_state=41)

et_classifier.fit(X_train, y_train)

prediction_etc = et_classifier.predict(X_test)

accuracy = accuracy_score(y_test, prediction_etc)
print(f"Accuracy: {accuracy}")
```

```python
prediction_etc =pd.DataFrame(prediction_etc, columns=["Predicted Churn2"])
```

```python
import xgboost as xgb

dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

params = {
    'max_depth': 3, 
    'eta': 0.1, 
    'objective': 'binary:logistic', 
    'eval_metric': 'error' 
}

num_round = 100 
xgb_model = xgb.train(params, dtrain, num_boost_round=num_round)

# Make predictions
preds = xgb_model.predict(dtest)
prediction_xgb = [1 if x > 0.5 else 0 for x in preds] 

# Evaluate the model
accuracy = accuracy_score(y_test, prediction_xgb)
print(f"Accuracy: {accuracy}")
```
